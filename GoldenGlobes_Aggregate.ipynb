```python
import pandas as pd
from bs4 import BeautifulSoup
import qgrid
```

<h3>Load imdb data from excel</h3>

```python
df = pd.read_excel('IMDB_data_GoldenGlobes.xlsx')

qgrid.show_grid(df)
```

<h3>Movies / TV Series with most nominations and wins</h3>

```python
df_movies = df.groupby(['Year', 'Movie']) \
    .agg(Nominations = ('Winner', "count"), Wins = ('Winner', sum)) \
    .sort_values(by = ['Wins', 'Nominations'], ascending = False) \
    .reset_index()

qgrid.show_grid(df_movies)
```

<h3>Persons with most wins</h3>

```python
df_persons = df.groupby(['Name']) \
    .agg(Nominations = ('Winner', "count"), Wins = ('Winner', sum)) \
    .sort_values(by = ['Wins', 'Nominations'], ascending = False) \
    .reset_index()

qgrid.show_grid(df_persons)
```

<h3>exclude some categories</h3>
no TV golden globes, Animated Films, Foreign Language Films, Documentaries, Debuts<br>
<b>start analysis in 1970</b>

```python
df_movies = df[~df.Category.str.contains('Television|TV|Series|Newcomer|New Star|Debut|Animated|Foreign|World|International|Documentary')]

#Start 1970
df_movies = df_movies.loc[df_movies.Year >= 1970]

df_movies.Category.unique()
#pd.DataFrame(df_movies.Category.unique()).to_excel("categories_globes.xlsx")
```

<h3>combine categories (historically, renaming has happend regularly)</h3>

```python
replaceValues = {
    'Best Actor in a Motion Picture - Comedy or Musical': 'Best Actor - Comedy or Musical'
    , 'Best Actor in a Motion Picture - Drama': 'Best Actor - Drama'
    , 'Best Actress in a Motion Picture - Comedy or Musical': 'Best Actress - Comedy or Musical'
    , 'Best Actress in a Motion Picture - Drama': 'Best Actress - Drama'
    , 'Best Original Score - Motion Picture': 'Best Original Score'
    , 'Best Original Song - Motion Picture': 'Best Original Song'
    , 'Best Original Song' : 'Best Song'
    , 'Best Screenplay - Motion Picture': 'Best Screenplay'
    , 'Best Supporting Actor - Motion Picture': 'Best Supporting Actor'
    , 'Best Supporting Actress - Motion Picture': 'Best Supporting Actress'
    , 'Best Director - Motion Picture': 'Best Director'
    , 'Best Actor in Motion Picture - Drama': 'Best Actor - Drama'
    , 'Best Actor in a Supporting Role - Motion Picture': 'Best Supporting Actor'
    , 'Best Actress in a Supporting Role - Motion Picture': 'Best Supporting Actress'
    , 'Best Performance by an Actor in a Motion Picture - Comedy or Musical': 'Best Actor - Comedy or Musical'
    , 'Best Performance by an Actor in a Motion Picture - Drama': 'Best Actor - Drama'
    , 'Best Performance by an Actor in a Supporting Role in a Motion Picture': 'Best Supporting Actor'
    , 'Best Performance by an Actress in a Motion Picture - Comedy or Musical': 'Best Actress - Comedy or Musical'
    , 'Best Performance by an Actress in a Motion Picture - Drama': 'Best Actress - Drama'
    , 'Best Performance by an Actress in a Supporting Role in a Motion Picture': 'Best Supporting Actress'
    , 'Best Motion Picture - Musical or Comedy': 'Best Motion Picture - Comedy or Musical'
    , 'Best Performance by an Actor in a Motion Picture - Musical or Comedy': 'Best Actor - Comedy or Musical'
    , 'Best Performance by an Actress in a Motion Picture - Musical or Comedy': 'Best Actress - Comedy or Musical'
    , 'Best Motion Picture - Comedy' : 'Best Motion Picture - Comedy or Musical'
    , 'Best Motion Picture - Musical' : 'Best Motion Picture - Comedy or Musical'
    , 'Best Cinematography - Black and White' : 'Best Cinematography'
    , 'Best Cinematography - Color' : 'Best Cinematography'
}

df_movies.loc[:, 'Category'] = df_movies.loc[:, 'Category'].replace(replaceValues)
df_movies.Category.unique()
```

```python
import qgrid
qgrid.show_grid(df_movies, show_toolbar=False)
```

<h3>Pivot Table: Prepare Data for modelling</h3>

```python
#1 = nominee
#2 = winner
#0 = no nominee, no winner

#pivot table returns coding -1 / 0 / 1 --> +1 gives desired results

df_winners = df_movies.pivot_table(index = ['Year', 'Movie'], columns = 'Category', values = 'Winner', aggfunc = "max") \
    .fillna(-1) \
    .apply(lambda x: x + 1) \
    .reset_index()

#Add Golden Globes to Column name
columns_old = df_winners.columns.values
columns_old[2:] = 'GLOBES ' + columns_old[2:]
df_winners.columns = columns_old

df_winners
```
